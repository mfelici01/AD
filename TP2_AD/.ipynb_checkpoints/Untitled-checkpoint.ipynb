```python
# Calcul des corrélations entre les variables initiales et les composantes principales
corr_matrix = np.dot(acp.components_.T, np.diag(np.sqrt(acp.explained_variance_)))

```

```python

# Visualisation des corrélations entre les matières initiales et les trois premiers axes
bar_width = 0.2  # Largeur des barres

plt.bar(np.arange(len(nomv)), corr_matrix[0, :], width=bar_width, label='PC1')
plt.bar(np.arange(len(nomv)) + bar_width, corr_matrix[1, :], width=bar_width, label='PC2')
plt.bar(np.arange(len(nomv)) + 2 * bar_width, corr_matrix[2, :], width=bar_width, label='PC3')

plt.xlabel('Matières')
plt.ylabel('Corrélation')
plt.title('Corrélation entre les matières et les trois premiers axes')
plt.xticks(np.arange(len(nomv)) + bar_width, nomv)  # Ajuster les étiquettes d'axe x
plt.legend()
plt.show()
```
