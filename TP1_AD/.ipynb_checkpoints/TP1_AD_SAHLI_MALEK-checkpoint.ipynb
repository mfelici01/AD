```python
# Générer des données pour l'exemple
data, _ = make_blobs(n_samples=300, centers=2, random_state=42)

methods = ['single', 'complete', 'average', 'ward']
linkages = [CAH(data, method=method) for method in methods]
# Calculer les étiquettes de clustering pour chaque méthode
t = 2  # 2 clusters
labels_single = fcluster(linkages[0], t=t, criterion='maxclust')
labels_complete = fcluster(linkages[1], t=t, criterion='maxclust')
labels_average = fcluster(linkages[2], t=t, criterion='maxclust')
labels_ward = fcluster(linkages[3], t=t, criterion='maxclust')

# Calculer le coefficient de silhouette pour chaque méthode
silhouette_single = silhouette_score(data, labels_single)
silhouette_complete = silhouette_score(data, labels_complete)
silhouette_average = silhouette_score(data, labels_average)
silhouette_ward = silhouette_score(data, labels_ward)

# Afficher les scores de silhouette
print("\nSilhouette Scores:")
print(f"Méthode Single : {silhouette_single}")
print(f"Méthode Complete : {silhouette_complete}")
print(f"Méthode Average : {silhouette_average}")
print(f"Méthode Ward : {silhouette_ward}")

# Afficher les dendrogrammes sur le même graphique
plt.figure(figsize=(15, 8))
for i, method in enumerate(methods):
    plt.subplot(2, 2, i + 1)
    dendrogram(linkages[i])
    plt.title(f'Dendrogramme - Méthode: {method}')

plt.tight_layout()
plt.show()
```

 la principale différence entre ces méthodes réside dans la façon dont elles mesurent la similarité entre les clusters.
 on peut voir que la methode single forme des clusters allongés, car elle est sensible aux valeurs anormal.
la methode average produit des clusters plus équilibrés que celui de la methode single .
la methode complete forme des clusters compacts.
la methode ward forme des clusters plus régulière et équilibrée

<p>2.1 Utilisation de la methode K-Means pour la reduction de couleur</p>

```python
# Charger l'image
img = np.float32(mpimg.imread('visage.bmp'))

# Organiser les données de l'image couleur en tableau 256x256x3
data = np.reshape(img, (256*256, 3))

# Définir le nombre de couleurs
K = 100

# Appliquer K-Means
kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
kmeans.fit(data)

# Étape de codage-décodage
compressed_data = kmeans.cluster_centers_[kmeans.labels_]
result_img = np.reshape(compressed_data, (256, 256, 3))

# Visualiser l'image resultat
plt.imshow(result_img/255)
plt.show()
```

```python
for K in [2, 4, 8, 16]:
    # Appliquer K-Means
    kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
    kmeans.fit(data)

    # Étape de codage-décodage
    compressed_data = kmeans.cluster_centers_[kmeans.labels_]
    result_img = np.reshape(compressed_data, (256, 256, 3))

    # Visualiser l'image résultante
    plt.imshow(result_img/255)
    plt.title(f'Compression with K={K}')
    plt.show()
```

```python
distortions = []
for K in range(2, 20):
    kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
    kmeans.fit(data)
    distortions.append(kmeans.inertia_)

# Visualiser la distorsion en fonction de K
plt.plot(range(2, 20), distortions, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Distortion')
plt.title('Elbow Method K')
plt.show()
```

cette methode aide a identifier le point auquel la somme des carres intra-cluster commence à diminuer à un rythme plus lent. 
La valeur optimale de K est au niveau du coude, car elle représente un bon équilibre entre 
la capture de clusters significatifs et la prévention du surajustement.

<p><strong>3 Realisation algorithmique</p>

<p>Q. 1 Developper l’algorithme de coalescence.</p>

```python
def coalescence(x, K, g):
    N = x.shape[1]
    clas = np.zeros(N)
    g2 = g.copy()

    max_iter = 100

    for _ in range(max_iter):
        for i in range(N):
            distances = np.sum((x[:, i] - g2.T) ** 2, axis=1)
            clas[i] = np.argmin(distances)

        for j in range(K):
            points_in_cluster = x[:, clas == j]
            if points_in_cluster.shape[1] > 0:
                g2[j, :] = np.mean(points_in_cluster, axis=1)

    return clas, g2
```

x: Une matrice où chaque colonne représente un individu avec deux mesures (caractéristiques) pour chaque individu.
K: Le nombre de classes souhaité.
g: Les centres de gravité initiaux.
N est le nombre d'individus, initialisation des vecteurs de classe clas et des centres de gravité g2.

La boucle principale s'exécute pour un nombre maximal d'itérations (max_iter) afin de trouver la convergence.

À l'intérieur de la boucle, les individus sont attribués à des classes en fonction des centres de gravité actuels, puis les centres de gravité sont mis à jour en fonction des nouvelles classes.

En fin de compte, la fonction renvoie les classes finales et les centres de gravité finaux.

<p>Q. 2 Appliquer l’algorithme de coalescence. Vous devez donc creer 2 centres de gravite initiaux
(parametres g de la fonction). Pour cela, vous allez choisir au hasard 2 individus de l’ensemble
d’apprentissage qui constitueront les centres initiaux</p>

```python
#Générer les données selon les distributions gaussiennes
np.random.seed(42)
mean1 = np.array([2, 2])
covariance1 = 2 * np.identity(2)
data1 = np.random.multivariate_normal(mean1, covariance1, 128).T

mean2 = np.array([-4, -4])
covariance2 = 6 * np.identity(2)
data2 = np.random.multivariate_normal(mean2, covariance2, 128).T

#Concaténer les données
x = np.concatenate((data1, data2), axis=1)

#Initialiser les centres de gravité en choisissant au hasard deux individus de l'ensemble d'apprentissage
initial_centers = x[:, np.random.choice(x.shape[1], size=2, replace=False)].T

#Appliquer la fonction coalescence
clas, final_centers = coalescence(x, 2, initial_centers)

#Afficher les résultats
print("Classes finales :", clas)
print("Centres de gravité finaux :\n", final_centers)
```

Deux ensembles de données (data1 et data2) sont générés à partir de distributions gaussiennes bidimensionnelles.
mean1 et mean2 sont les moyennes des distributions.
covariance1 et covariance2 sont les matrices de covariance des distributions.
Les données sont générées à l'aide de np.random.multivariate_normal.
Les deux ensembles de données sont concaténés le long de l'axe des colonnes pour former un ensemble de données unique (x).
Deux centres de gravité initiaux sont choisis au hasard parmi les individus de l'ensemble d'apprentissage.
La fonction coalescence est appelée avec les données (x), le nombre de classes désiré (2), et les centres de gravité initiaux.
La fonction retourne les classes attribuées à chaque individu (clas) et les centres de gravité finaux (final_centers).

Les résultats de la partition montrent que tous les individus ont été attribués à la classe 0, à l'exception d'un seul individu qui a été attribué à la classe 1. Les centres de gravité finaux sont situés à environ 
[−0.55,0.02] pour la classe 0 et 
[−3.21,−6.59] pour la classe 1.
